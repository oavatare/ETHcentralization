```python
# https://github.com/ethereum/wiki/wiki/JavaScript-API
# https://web3js.readthedocs.io/en/1.0/index.html
import pandas as pd
from collections import Counter
from datetime import datetime
minerCount = []
# Can run an analysis of mining centralization over the entire history of Ethereum
for i in range(web3.eth.blockNumber - 50, web3.eth.blockNumber):#web3.eth.blockNumber):
    thisETHblock = web3.eth.getBlock(i)
    #print(thisETHblock.miner)
    minerCount.append(thisETHblock.miner)
    #print(datetime.utcfromtimestamp(int(thisETHblock.timestamp)).strftime('%Y-%m-%d %H:%M:%S'))
c = Counter(minerCount)
df = pd.DataFrame.from_dict(c, orient='index').reset_index()
df = df.rename(columns={'index':'Miner', 0:'Count'})
df["Frequency "] = round(df['Count'] / len(minerCount), 2) 
df = df.sort_values(by='Count', ascending=False)
df
```

```python
w, h = 4, len(df);
minerBalance = [[0 for x in range(w)] for y in range(h)] 
#print(minerBalance)

import requests
r = requests.get('https://api.cryptowat.ch/markets/coinbase-pro/ethusd/price')
r = r.json()
price = r['result']['price']
print("Ethereum price: ", price)

for i in range(0, len(df)):
    minerBalance[i][0] = df['Miner'].iloc[i]
    minerBalance[i][1] = df['Count'].iloc[i] / df['Count'].sum()
    minerBalance[i][2] = int(web3.eth.getBalance(df['Miner'].iloc[i]) / 1000000000000000000)
    minerBalance[i][3] = int(web3.eth.getBalance(df['Miner'].iloc[i]) / 1000000000000000000) * price
minerBalance = pd.DataFrame(minerBalance)
minerBalance.columns = ['Miner', 'PercentageOfEthereumMined', 'ETH_Amount', 'ETH_Value_In_USD']
minerBalance.sort_values("ETH_Amount", inplace=True, ascending=False)
minerBalance
```
